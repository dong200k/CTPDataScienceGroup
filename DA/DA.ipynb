```python
# Our data management libraries
import pandas as pd
import numpy as np

# A basic visualization library
import matplotlib.pyplot as plt

# A great visualization library
import seaborn as sns

# This module provides regular expression matching operations similar to those found in Perl.
import re

# Very important, this will make your charts appear in your notebook instead of in a new window.
%matplotlib inline
```

```python
# import data 
df = pd.read_csv('data/DS_DA_BS_Gene.csv')
df.head()
```

```python
# In the column names, replace all the spaces with an underscore, and make them all lowercase as well
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = str.lower)
df.head()
```

```python
# check duplicates
df.duplicated().sum()
```

```python
# clean the salary data
salary_est=list(df['salary_estimate'])
salary_re=[]

for i in salary_est:
    a=re.findall(r"\d+",i)
    salary_re.append(a)
  
df['salary_cleaned']=salary_re
df.head()
```

```python
# clean the revenue data
revenue_list=list(df['revenue'])
revenue_array = []

for i in revenue_list:
    if "Unknown / Non-Applicable" in i:
        revenue_array.append(-1)
    elif "-1" in i:
        revenue_array.append(-1)
    elif "$10+" in i:
        revenue_array.append(10000000)
    elif "Less than $1" in i:
        revenue_array.append(100)
    else: 
        a = i.rfind("to ")
        newI = i[a + 4:]
        b = re.split(r"\s+",newI)
        
        if b[1] == "million":
            new_num = int(b[0])*1000
            revenue_array.append(new_num)
        elif b[1] == "billion":
            new_num = int(b[0])*1000000
            revenue_array.append(new_num)

df['revenue_cleaned']=revenue_array
df.head() 
```

```python
# add min and max salary col
salary_min=[]
salary_max=[]
for i in salary_re:
    if len(i)==2:
        salary_min.append(int(i[0]))
        salary_max.append(int(i[1]))
    if len(i)==1:
        salary_min.append(int(i[0]))
        salary_max.append(int(i[0]))
df['min_salary']=salary_min
df['max_salary']=salary_max
df.head()
```

```python
# add average salary column
salary_ave = []
for i in range(len(salary_min)):
    ave = (salary_min[i] + salary_max[i])/2
    salary_ave.append(int(ave))
    
df['average_salary']=salary_ave
```

```python
Job_counts = df['job_title'].value_counts()
Job_counts.head()
```

```python
# remove -1: no valid data from revenue_cleaned column
condition1 = df['revenue_cleaned'] > 0

# Pick job title as Data Analyst
condition2 = df['job_title'] == 'Data Analyst'

Data = df[condition1 & condition2]
Data[['job_title','revenue_cleaned', 'average_salary']]
```

# Which sector provides the highest salary for DS/BA/DA?

```python
df[['sector', 'average_salary']]
```

```python
# group by sector
cond = df['sector'] != '-1'
df = df[cond]
gb = df.groupby('sector')

# find the mean value in the same sector and sort it in ascending order.
average_salary_mean = gb['average_salary'].mean().sort_values(ascending=True)

average_salary_mean.plot(kind='barh', figsize=(15,9))
print("Biotech & Pharmaceuticals sector provider highest salary in all related DA jobs.")
```
